# Extract overall Survival rate

```python

gb=df.groupby('Survived').agg({'Survived':'count'})
gb.index=['Survived','Died']
gb


```

```python
# Survival counts
survival_counts = df['Survived'].value_counts()

# Labels
labels = ['Not Survived', 'Survived']

# Apply seaborn styling
sns.set_palette("pastel")     # soft colors
sns.set_style("whitegrid")    # clean style

# Plot
plt.figure(figsize=(6, 6))
plt.pie(
    survival_counts,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90,
    wedgeprops={'edgecolor': 'black'}
)

plt.title("Overall Survival Rate")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.set_context("poster")
sns.set_theme("poster")
sns.set_style("ticks")
palette=sns.set_palette("Set1")
sns.countplot(x='Survived',data=df,palette=palette,edgecolor='black')
plt.title("Overall Survival Count",fontsize=24,weight='bold')
plt.xlabel("Survival Status",weight='bold',fontsize=15)
plt.ylabel("Count of Passenger",weight='bold',fontsize=15)
plt.xticks([0,1],['Died','Survived'])
sns.despine(top=True,right=True)
plt.tight_layout()
```

### Survived Count â€” Analysis
- Non-survivors significantly outnumber survivors.
- This class imbalance is crucial to consider when training classification models.
